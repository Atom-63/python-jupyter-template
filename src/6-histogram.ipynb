### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import numpy as np
import matplotlib.pyplot as plt

# Create a figure with a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Line Chart (Top-Left)
x = np.linspace(0, 10, 100)
y = np.sin(x)
axes[0, 0].plot(x, y, color='blue')
axes[0, 0].set_title("Line Chart")
axes[0, 0].set_xlabel("X values")
axes[0, 0].set_ylabel("Y values")

# Bar Chart (Top-Right)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
axes[0, 1].set_title("Bar Chart")
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")

# Scatter Plot (Bottom-Left)
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x_scatter, y_scatter, color='blue')
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("X values")
axes[1, 0].set_ylabel("Y values")

# Pie Chart (Bottom-Right)
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'red', 'green', 'purple'])
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Create a new figure for the histogram
plt.figure(figsize=(6, 4))

# Histogram Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
plt.hist(data, bins=5, color='cyan', edgecolor='black')
plt.title("Histogram of Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plots
plt.show()
```
